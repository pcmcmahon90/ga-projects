# Metacritic webscraper

The goal was to capture metacritic scores for all movies on the metacritic website. Metacritic scores represent primarily critic reviews, with a small weight torwards user scores and some other factors. This webscraper was built to go page by page through the historic catalog of metacritic films, and mine their data. 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd
import numpy as np
import urllib.request
import re

class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/5.0"
    
from bs4 import BeautifulSoup
```

```python
df = pd.DataFrame(columns = ['title','metacritic_score'])

# Scan from page 0 to page 111 of the metacritic 'browse all movies'.
# The scraper will go page by page, identify a starting location based on the html tag 'td' for class 'rank_wrapper thin'
# Then will grab the appropriate elements, add to a dataframe, then loop to the next page
          # Note : Certainly not the most elegant navigation of beautiful soup. But it did work.

# A sample of 3 pages was re-ran below for demonstration of the output:

for page in range(0,112):
    print("---Page "+str(page)+" scrape start:")
    scores = []
    titles = []
    df_temp = pd.DataFrame(columns = ['title','metacritic_score'])
    opener = AppURLopener()
    req = opener.open('http://www.metacritic.com/browse/movies/score/metascore/all/filtered?page='+str(page))
    soup = BeautifulSoup(req)
    x = soup.body
    tdlist = x.findAll('td', {"class": "rank_wrapper thin"})
    for i in tdlist:
        scores.append(i.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element)
        titles.append(i.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.find('a').text)
    if len(scores) == len(titles):
        df_temp['metacritic_score'] = scores
        df_temp['title'] = titles
        df = df.append(df_temp)
    else:
        print("List size failure at page:"+str(page))
```
