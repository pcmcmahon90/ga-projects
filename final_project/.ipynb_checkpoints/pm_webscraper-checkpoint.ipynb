```python
import pandas as pd
import numpy as np
import urllib.request
import re

class AppURLopener(urllib.request.FancyURLopener):
    version = "Mozilla/5.0"
    
from bs4 import BeautifulSoup
```

```python
opener = AppURLopener()
req = opener.open('http://www.metacritic.com/browse/movies/score/metascore/all/filtered?page=1')
```

```python
soup = BeautifulSoup(req)
```

```python
x = soup.body
x
```

```python
tdlist = x.findAll('td', {"class": "rank_wrapper thin"})
```

```python
df = pd.DataFrame(columns = ['title','metacritic_score','user_score'])

scores = []
titles = []
userscores = []
for i in tdlist:

    scores.append(i.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element)
    titles.append(i.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.next_element.find('a').text)
    userscores.append(i.find_next('div', {"class": "userscore_text"}).next_element.next_element.next_element.next_element.next_element.text)
df['user_score'] = userscores
df['metacritic_score'] = scores
df['title'] = titles
```
