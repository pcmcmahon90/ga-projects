```python
# Remove other y values from df for now, only focusing on total cume

df = df.drop(["Unnamed: 0","Total DVD/Video Release Window", "Total Locs At Opening", "Total Locs At Widest Release", "Total Opening 7 Day Gross", "Total Opening Wknd Avg","Total Opening Day Gross","Total Opening Wknd Gross","Total Opening Wknd Locs","Total Run Time"], axis = 1)
```

```python
df.info()
```

```python
df = df.select_dtypes(include='int64')
df
```

```python
# Set up X, y
X = df.drop('Total Cume', axis = 1)
y = df['Total Cume']
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# Define models

lgr = LogisticRegression()
nbclf = GaussianNB()
rf_clf = RandomForestClassifier(n_estimators=10)
knn = KNeighborsClassifier(n_neighbors=5)
voting_clf = VotingClassifier(estimators = [('rf', rf_clf), ('lr',lgr), ('nb', nbclf), ('knn', knn)], voting = 'hard')

# Voting

voting_clf.fit(X_train, y_train)

preds = voting_clf.predict(X_test)
print(classification_report(preds, y_test))
```
